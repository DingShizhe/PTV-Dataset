```python
import numpy as np
import os

ptv_data_dir = "/home/dingshizhe/mnt/INTER/jet/field_data/"
fns = sorted(list(os.listdir(ptv_data_dir)))
fns = [os.path.join(ptv_data_dir, fn) for fn in fns]
```

```python
data = [np.load(fn) for fn in fns]

ls = [len(d) for d in data]
print(ls)
```

```python
ls = [ x for x in ls if x > 2000]
print(max(ls))
```

```python
import matplotlib.pyplot as plt

plt.hist(ls, bins=100, alpha=0.75)
plt.show()
```

```python
ls = np.array(ls)

print(np.argmin(ls))
print(np.min(ls))
```

```python
top_100 = ls.argsort()[:100]

print(top_100)
print(ls[top_100])
```
